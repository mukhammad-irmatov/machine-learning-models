```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression  
```

```python
df = pd.read_csv("townHomeprices.csv")
df
```

```python
pd.get_dummies(df.town)
dummies = pd.get_dummies(df.town)
```

```python
dummies
```

```python
merged = pd.concat([df, dummies], axis = 1)
merged
```

```python
final = merged.drop(['town','nursultan'], axis = 1)
final
```

```python
model = LinearRegression()
X = final.drop(['price'], axis = 'columns')
X
```

```python
y = final.price
y 
```

```python
model.fit(X,y)
```

```python
model.predict([[3600, 0, 0]])
```

```python
model.predict([[3400, 0, 1]])
```

```python
model.score(X,y)
```
