```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
```

```python
df = pd.read_csv("carprices3.csv")
df
```

```python
pd.get_dummies(df.Car_Model)
dummies = pd.get_dummies(df.Car_Model)
```

```python
dummies
```

```python
merged = pd.concat([df, dummies], axis = 1)
merged
```

```python
final = merged.drop(['Car_Model','Mercedez Benz C class'], axis = 1)
final
```

```python
model = LinearRegression()
X = final.drop(['Sell Price($)'], axis = 'columns')
X
```

```python
y = final["Sell Price($)"]
y 
```

```python
model.fit(X,y)
```

```python
model.predict([[45000, 5, 0,0]])
```

```python
model.predict([[30000, 3, 0,0]])
```

```python
model.score(X,y)
```
